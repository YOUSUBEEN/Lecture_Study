생활인구합계가 1000이상이고, 행정동 코드가 11680610인 데이터만 뽑아내자(and)

```python
condition = (sample["생활인구합계"] >= 1000) & (sample["행정동코드"] == 11680610)
condition
```

```python
sample[condition]
```

```python
# OR방법 (~이거나)

condition = (sample["생활인구합계"] >= 1000) | (sample["행정동코드"] == 11680610)
condition
```

apply함수
- dataframe에 python함수를 적용할 수 있음
- 예를들어, 가로-세로-높이를 이용해 부피라는 컬럼을 추가해보자.

```python
df
```

```python
def getVolume(row):
    return row["가로"] * row["세로"] * row["높이"]

df["부피"] = df.apply(getVolume,axis=1)
df
```

중복 데이터 처리
- 중복 데이터 찾기 : duplicated

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'lndomie', 'lndomie','lndomie'],
    'style': ['cup', 'cup','cup','pack','pack'],
    'rating':[4, 4, 3.5 ,15 ,5]
})

df
```

```python
df.duplicated()
```

```python
# Keep이라는 속성을 사용하여 어떤 값을 중복으로 인식할 것인지 설정, 기본값은 first
# 중복된 값을 첫 번째로.
df.duplicated(keep="last")
```

```python
# subset 속성을 사용하여 특정 컬럼에 대한 중복만 처리 가능
df.duplicated(subset=["brand"])
```

중복값 삭제 : drop.duplicates

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset=['brand','style'], keep='last')
```

결측(빈) 데이터 처리
- isnull(isna) 결측치 찾기

```python
import numpy as np
```

```python
df = pd.DataFrame([[1,5,2],[np.nan,1,3], [6,np.nan,20],[2,5,np.nan],[2,5,2]])
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

fillna : 결측치 채우기

```python
df
```

```python
df.fillna(1) # 특정 값으로 NaN부분 채우기
```

```python
# 특정 값으로 채우기 : ffill = forward fill 앞에 값으로 대체
df.fillna(method="ffill")
```

```python
# 특정 값으로 채우기 : bfill = backward fill 앞에 값으로 대체
df.fillna(method="bfill")
```
